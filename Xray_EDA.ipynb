### 画像の可視化

```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image

from glob import glob
```

```python
# 画像を表示する関数
def visualize_images(num_imgs, phase='train'):
    plt.figure(figsize=(8, 4*num_imgs))
    for n in range(num_imgs):

        # 入力用の画像データ
        if phase == 'train':

            # 入力用の画像データ
            image = Image.open(f'Segmentation01/{phase}/org/{n+1}.png')
            image = np.array(image)

            # 出力用の画像データ
            label = Image.open(f'Segmentation01/{phase}/label/{n+1}.png')
            label = np.array(label)

        else:
            # 入力用の画像データ
            image = Image.open(f'Segmentation01/{phase}/org/{n}.png')
            image = np.array(image)

            # 出力用の画像データ
            label = Image.open(f'Segmentation01/{phase}/label/{n}.png')
            label = np.array(label)

        plt.subplot(num_imgs, 2, 2*n+1)
        plt.imshow(image, cmap='gray')

        plt.subplot(num_imgs, 2, 2*n+2)
        plt.imshow(label, cmap='gray')
```

```python
# 画像の表示
visualize_images(5, 'test')
```
